# üîÄ If, Elif, Else - Estructuras de Control

En este notebook dominaremos las estructuras de control condicionales, la base de la l√≥gica de programaci√≥n que permite a nuestros programas tomar decisiones inteligentes.

## ‚úÖ Objetivos
- Dominar if, elif, else
- Usar operadores l√≥gicos
- Crear condiciones complejas

## 1. Estructura If B√°sica

La estructura `if` nos permite ejecutar c√≥digo solo cuando se cumple una condici√≥n:

```python
# Ejemplo b√°sico de if
edad = 25

if edad >= 18:
    print("‚úÖ Eres mayor de edad")
    print("Puedes votar y conducir")

print("Este mensaje siempre se ejecuta")

# Ejemplo con datos de empleado
salario = 50000
experiencia = 3

if salario > 45000:
    print(f"üí∞ Salario competitivo: ${salario:,}")

if experiencia >= 2:
    print(f"üë®‚Äçüíº Empleado experimentado: {experiencia} a√±os")
```

## 2. Estructura If-Else

Con `else` podemos ejecutar c√≥digo alternativo cuando la condici√≥n es falsa:

```python
# Sistema de acceso b√°sico
usuario = "admin"
password = "123456"

if usuario == "admin" and password == "secreto":
    print("üîì Acceso concedido")
    print("Bienvenido al sistema")
else:
    print("üîí Acceso denegado")
    print("Usuario o contrase√±a incorrectos")

# Evaluaci√≥n de rendimiento
ventas_mes = 85000
objetivo = 80000

if ventas_mes >= objetivo:
    cumplimiento = (ventas_mes / objetivo) * 100
    print(f"üéØ Objetivo cumplido: {cumplimiento:.1f}%")
    print(f"Ventas: ${ventas_mes:,} / Objetivo: ${objetivo:,}")
else:
    faltante = objetivo - ventas_mes
    print(f"‚ö†Ô∏è Objetivo no alcanzado")
    print(f"Faltaron: ${faltante:,}")
```

## 3. Estructura If-Elif-Else

Para m√∫ltiples condiciones usamos `elif` (else if):

```python
# Sistema de calificaciones
calificacion = 87

if calificacion >= 90:
    letra = "A"
    estado = "Excelente"
elif calificacion >= 80:
    letra = "B"
    estado = "Bueno"
elif calificacion >= 70:
    letra = "C"
    estado = "Satisfactorio"
elif calificacion >= 60:
    letra = "D"
    estado = "Suficiente"
else:
    letra = "F"
    estado = "Reprobado"

print(f"Calificaci√≥n: {calificacion}")
print(f"Letra: {letra}")
print(f"Estado: {estado}")
```

## 4. Operadores de Comparaci√≥n

```python
# Todos los operadores de comparaci√≥n
a = 10
b = 5
c = 10

print("Operadores de comparaci√≥n:")
print(f"{a} == {c}: {a == c}")    # Igual
print(f"{a} != {b}: {a != b}")    # Diferente
print(f"{a} > {b}: {a > b}")      # Mayor que
print(f"{a} < {b}: {a < b}")      # Menor que
print(f"{a} >= {c}: {a >= c}")    # Mayor o igual
print(f"{b} <= {a}: {b <= a}")    # Menor o igual

# Comparaciones con strings
nombre1 = "Ana"
nombre2 = "Carlos"

print(f"\nComparaciones de strings:")
print(f"'{nombre1}' == '{nombre2}': {nombre1 == nombre2}")
print(f"'{nombre1}' < '{nombre2}': {nombre1 < nombre2}")  # Orden alfab√©tico

# Operador in para verificar pertenencia
frutas = ["manzana", "pera", "uva"]
print(f"\n'manzana' in frutas: {'manzana' in frutas}")
print(f"'pl√°tano' in frutas: {'pl√°tano' in frutas}")
```

## 5. Operadores L√≥gicos

Combinamos condiciones con `and`, `or`, `not`:

```python
# Operador AND - todas las condiciones deben ser verdaderas
edad = 25
tiene_licencia = True
tiene_auto = True

if edad >= 18 and tiene_licencia and tiene_auto:
    print("‚úÖ Puede conducir legalmente")
else:
    print("‚ùå No puede conducir")

# Operador OR - al menos una condici√≥n debe ser verdadera
es_estudiante = False
es_senior = True
es_empleado = False

if es_estudiante or es_senior or es_empleado:
    print("üé´ Califica para descuento")
else:
    print("üí∞ Precio regular")

# Operador NOT - invierte el valor l√≥gico
sistema_activo = True

if not sistema_activo:
    print("üî¥ Sistema fuera de servicio")
else:
    print("üü¢ Sistema operativo")
```

## 6. üéØ Ejercicios Pr√°cticos

### Ejercicio 1: Sistema de Pr√©stamos Bancarios

```python
# TODO: Sistema de evaluaci√≥n de pr√©stamos
print("üè¶ SISTEMA DE EVALUACI√ìN DE PR√âSTAMOS")
print("=" * 45)

# Datos del solicitante
nombre = "Mar√≠a Gonz√°lez"
edad = 32
ingresos_mensuales = 35000
a√±os_empleo = 3
score_crediticio = 720
monto_solicitado = 250000
tiene_aval = True

print(f"Solicitante: {nombre}")
print(f"Edad: {edad} a√±os")
print(f"Ingresos: ${ingresos_mensuales:,}")
print(f"A√±os de empleo: {a√±os_empleo}")
print(f"Score crediticio: {score_crediticio}")
print(f"Monto solicitado: ${monto_solicitado:,}")
print(f"Tiene aval: {'S√≠' if tiene_aval else 'No'}")
print()

# L√≥gica de aprobaci√≥n
aprobado = False
tasa_interes = 0
razon = ""

# Criterios b√°sicos
if edad < 18 or edad > 65:
    razon = "Edad fuera del rango permitido (18-65 a√±os)"
elif ingresos_mensuales < 15000:
    razon = "Ingresos insuficientes (m√≠nimo $15,000)"
elif a√±os_empleo < 1:
    razon = "Insuficiente estabilidad laboral (m√≠nimo 1 a√±o)"
elif score_crediticio < 600:
    razon = "Score crediticio bajo (m√≠nimo 600)"
elif monto_solicitado > (ingresos_mensuales * 60):  # M√°ximo 5 a√±os de ingresos
    razon = "Monto solicitado excede la capacidad de pago"
else:
    # Pr√©stamo pre-aprobado, ahora determinar la tasa
    aprobado = True
    
    if score_crediticio >= 800 and a√±os_empleo >= 5:
        tasa_interes = 8.5
        razon = "Excelente perfil crediticio"
    elif score_crediticio >= 750 and a√±os_empleo >= 3:
        tasa_interes = 9.5
        razon = "Buen perfil crediticio"
    elif score_crediticio >= 700 or tiene_aval:
        tasa_interes = 11.0
        razon = "Perfil crediticio aceptable"
    else:
        tasa_interes = 13.5
        razon = "Perfil crediticio con reservas"

# Mostrar resultado
print("RESULTADO DE LA EVALUACI√ìN:")
print("-" * 30)

if aprobado:
    pago_mensual = (monto_solicitado * (tasa_interes/100) / 12) / (1 - (1 + tasa_interes/100/12)**(-180))
    total_a_pagar = pago_mensual * 180  # 15 a√±os
    
    print("‚úÖ PR√âSTAMO APROBADO")
    print(f"Monto: ${monto_solicitado:,}")
    print(f"Tasa de inter√©s: {tasa_interes}% anual")
    print(f"Pago mensual: ${pago_mensual:,.2f}")
    print(f"Total a pagar: ${total_a_pagar:,.2f}")
    print(f"Raz√≥n: {razon}")
else:
    print("‚ùå PR√âSTAMO DENEGADO")
    print(f"Raz√≥n: {razon}")
```

### Ejercicio 2: Calculadora de Descuentos por Volumen

```python
# TODO: Sistema de descuentos empresariales
print("üõçÔ∏è SISTEMA DE DESCUENTOS POR VOLUMEN")
print("=" * 45)

# Datos de la compra
cliente = "Distribuidora L√≥pez SA"
es_cliente_vip = True
a√±os_cliente = 5
cantidad_productos = 150
precio_unitario = 85.50
es_temporada_alta = False

subtotal = cantidad_productos * precio_unitario

print(f"Cliente: {cliente}")
print(f"Cliente VIP: {'S√≠' if es_cliente_vip else 'No'}")
print(f"A√±os como cliente: {a√±os_cliente}")
print(f"Cantidad: {cantidad_productos} unidades")
print(f"Precio unitario: ${precio_unitario}")
print(f"Subtotal: ${subtotal:,.2f}")
print()

# L√≥gica de descuentos
descuento_volumen = 0
descuento_cliente = 0
descuento_fidelidad = 0

# Descuento por volumen
if cantidad_productos >= 200:
    descuento_volumen = 0.15  # 15%
elif cantidad_productos >= 100:
    descuento_volumen = 0.10  # 10%
elif cantidad_productos >= 50:
    descuento_volumen = 0.05  # 5%

# Descuento por tipo de cliente
if es_cliente_vip:
    descuento_cliente = 0.08  # 8% adicional para VIP

# Descuento por fidelidad (a√±os como cliente)
if a√±os_cliente >= 5:
    descuento_fidelidad = 0.05  # 5%
elif a√±os_cliente >= 3:
    descuento_fidelidad = 0.03  # 3%
elif a√±os_cliente >= 1:
    descuento_fidelidad = 0.01  # 1%

# Penalizaci√≥n por temporada alta
penalizacion_temporada = 0
if es_temporada_alta and not es_cliente_vip:
    penalizacion_temporada = 0.02  # Reducir descuentos en 2%

# Calcular descuento total
descuento_total = descuento_volumen + descuento_cliente + descuento_fidelidad - penalizacion_temporada

# L√≠mite m√°ximo de descuento
if descuento_total > 0.25:  # M√°ximo 25%
    descuento_total = 0.25

# C√°lculos finales
monto_descuento = subtotal * descuento_total
total_final = subtotal - monto_descuento

print("DESGLOSE DE DESCUENTOS:")
print("-" * 25)
if descuento_volumen > 0:
    print(f"Descuento por volumen: {descuento_volumen*100}%")
if descuento_cliente > 0:
    print(f"Descuento cliente VIP: {descuento_cliente*100}%")
if descuento_fidelidad > 0:
    print(f"Descuento fidelidad: {descuento_fidelidad*100}%")
if penalizacion_temporada > 0:
    print(f"Penalizaci√≥n temporada: -{penalizacion_temporada*100}%")

print(f"\nDESCUENTO TOTAL: {descuento_total*100}%")
print(f"Monto del descuento: ${monto_descuento:,.2f}")
print(f"üí∞ TOTAL A PAGAR: ${total_final:,.2f}")
print(f"Ahorro total: ${monto_descuento:,.2f}")
```
